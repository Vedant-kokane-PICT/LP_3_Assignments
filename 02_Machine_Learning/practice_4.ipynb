Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function y=(x+3)2 starting from the point x=2

find the local minima of the function y=(x+3)Â² starting from the point x=2

*  Equation = (x+3)^2
*   We need to find the local Minima using gradient descent

```python
# intial values
current_x = 2
learning_rate = 0.01
precision = 0.000001  # it is also called tolerance
max_iterations = 10000
```

```python
# slope will be the derivative of the equation using chain rule
# slope = 2 * (x + 3) (derivative using chain rule (dy/dx))
def slope(x):
  return 2 * (x + 3)

def y(x):
  return (x+3) ** 2
```

```python
y_values = []
x_values = []
x_values.append(current_x)
y_values.append(y(current_x))
```

```python
# In gradientdescent  new_x = old_x - (learning_rate * slope)

def gradient_descent(current_x, learning_rate, precision, max_iterations, x_values, y_values):
  iteration_count = 1
  while iteration_count <= max_iterations:
    previous_x = current_x
    current_x = current_x - (learning_rate * slope(current_x))
    x_values.append(current_x)
    y_values.append(y(current_x))
    if abs(current_x - previous_x) < precision:
      break

    print(f"Iteration - {iteration_count}" )
    print(f"X = {current_x}")
    iteration_count = iteration_count + 1
```

```python
gradient_descent(current_x,learning_rate,precision,max_iterations,x_values,y_values)
```

```python
# plotting the values of x and y on graph
import matplotlib.pyplot as plt
plt.scatter(x_values,y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent for equation - y=(x+3)^2')
plt.show()
```
